# Salvando graficos resultados de analises

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
plt.rc("figure",figsize=(14,6))
```

```python
dados = pd.read_csv("../output/dados_residenciais_tratados_v2outliers.csv",sep=";")
dados.head(10)
```

```python
#criar uma area para salvar os graficos
area = plt.figure()
```

```python
#criando os pontos das areas onde ficam os graficos
g1 = area.add_subplot(2, 2, 1)
g2 = area.add_subplot(2, 2, 2)
g3 = area.add_subplot(2, 2, 3)
g4 = area.add_subplot(2, 2, 4)
```

```python
g1.scatter(dados["Valor"], dados["Area"])
g1.set_title("Valor X  Area")

g2.hist(dados["Valor"])
g2.set_title("Histograma de valor")

g3.set_title("Amostra aleatoria de valor")
dados_g3 = dados["Valor"].sample(100)
dados_g3.index = range(dados_g3.shape[0])
g3.plot(dados_g3)

g4.set_title("Valor medio por tipo")
grupo = dados.groupby("Tipo")["Valor"]
media = grupo.mean().values
label = grupo.mean().index
g4.bar(label, media)
```

```python
#o bbox, que remove a borda branca da área do gráfico, tornando nossa apresentação mais elegante
area.savefig("../output/graphs.png", dpi=300, bbox_inches = 'tight')
```
